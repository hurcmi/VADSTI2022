```python
##-----------------------
## Required Task: 9 of 15
##-----------------------

# Regression -- Create the regression models and regression summary tables on slides 14, 15, 16, and 17

# Because the study is non-randomized, there is potential for selection bias with respect to treatment drug.
# If the study was randomized, we could just do a simple t-test comparing Drug A with Drug B.
# Linear regression model includes potential confounders:  Are Drug A patients older than Drug B patients?
#
#Age, sex and race typically included in regression model, but other factors can also be included (e.g., comorbidities).
#
#Regression model:
#          dBP ~ Drug + Age + Sex + Drug*Age + Drug*Sex 

        
        
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm

# Slide 15
# Generalized Linear Models 
# https://www.statsmodels.org/v0.10.1/glm.html
# https://www.statsmodels.org/stable/regression.html
# https://regenerativetoday.com/logistic-regression-model-analysis-visualization-and-prediction/

# The data
dataset = pd.read_excel("./HowardData.xls", sheet_name=0)
dataset.columns = ["dBP", "Drug", "Age", "Sex"]
dataset.head(n=3)

## Example1
formular1 = "dBP ~ Drug + Age + Sex + Drug*Age + Drug*Sex"
formular2 = "dBP ~ Drug + Age + Sex + Drug*Age"

#model = sm.GLM.from_formula(formular, family = sm.families.Binomial(), data=dataset)
model = sm.OLS.from_formula(formular1, data=dataset)
result = model.fit()
print(result.summary())

#-------------------
## Example2
#model = sm.GLM.from_formula("CD ~ Age", family = sm.families.Binomial(), data=dataset)
#result = model.fit()
#print(result.summary())

```

```python
##-----------------------
## Required Task: 11 of 15  *** Professor's Assignment ***
##-----------------------

#a: Recreate the table on slide 10 of HowardU-4-1-22.pptx
#b: Do the assignment at the botton of slide 10
#    => Calculate the p value and 95% confidence interval of the odds ratio for the table above.


## Odds And Log Odds  Slide 10 -- HowardU-4-1-22
# https://regenerativetoday.com/logistic-regression-model-analysis-visualization-and-prediction/
#
import math
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

#Using the observation on slide 39, compute the z-test of diffences in proportions for the 2 samples

#                 A    B
count = np.array([46,  54])
nobs = np.array([37,  63])

#table
table = [count, nobs]
display(table)

#dataframe
df = pd.DataFrame(table, columns=["A", "B"], index={'Intervention', 'Control'})
display(df)

#ztest
stat, pval = proportions_ztest(count, nobs)
print("\nz = %.4f      |    p = %.4f (two-sided)\n---\n\n" % (stat, pval))

#odds ratio
import scipy.stats as stats
oddsratio, pvalue = stats.fisher_exact(table)
print(f'Odds Ratio: {oddsratio:.2f}     |   p-Value: {pvalue:.2f}')

#Assignment:  Calculate the p value and 95% confidence interval of the odds ratio for the table above.
# see this: https://www.statology.org/confidence-interval-for-odds-ratio/
odds_ratio = (46*63) / (37*54)
odds_ratio_sqrt = 1.96 * np.sqrt(odds_ratio)

#
#print(math.e);  #print(math.exp(1))
Lower_95CI, Upper_95CI =  [ math.exp(1) ** (ln(.599) - odds_ratio_sqrt), math.exp(1) ** (ln(.599) + odds_ratio_sqrt) ]
print(f'\n--\nLower 95% CI = {Lower_95CI:.3f}   |  Upper 95% CI = {Upper_95CI:.3f}\n')
```

```python
##-----------------------
## Required Task: 14 of 15    ****   Challenge ***  Detailed Regression Fitting
##-----------------------

# Fitting Regression: Of X and y -- Follow along the code below and document the structure
#                                   of the data types of data used in cell

#--------
# Step 1: load libraries
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LinearRegression

#--------
# Step 2: load the data
X = np.arange(20) + np.random.normal(0,1,20)
X = X.reshape(-1, 1)
y = 2*X + 1

print(X[:3])
print(y[:3])

#--------
#Step 3: define model
model = LinearRegression()
print("Model type: ", model)

#--------
#Step 4: fit the model 
model.fit(X, y)

#--------
# Step 5: Make prediction
y_pred = model.predict(X)

#--------
# Step 6: Display True vs Predicted values
myList = {"Original_y":y[:,0], "Predicted_y":y_pred[:,0]}
df_prediction = pd.DataFrame(myList)
df_prediction.head()

#--------
# Step 7: Print accuracy metrices
from sklearn.metrics import mean_squared_error
import math
# calculate root mean squared error
testScore = math.sqrt(mean_squared_error(y, y_pred))
print('Test Score: %.2f RMSE' % (testScore))

#R-Square Score
from sklearn.metrics import r2_score
df_prediction = df_prediction.astype(float)
testScore = (r2_score(df_prediction.Original_y.values, df_prediction.Predicted_y))
print('Test Score: %.2f R2' % (testScore))

#--------
# step 8: Visualize True and predicted valies
plt.figure(figsize=(15, 7))
plt.title('True vs Predicted plot -- Linear Regression')
y_pred = y_pred + 0.5
plt.plot(y, label='Actual values')
plt.plot(y_pred,color='red',label='Predicted')
plt.legend()

plt.grid(True)
plt.show()


#--------
# Step 9: Define new data
Xnew = [0.1, 2,3,  3.4, 4.5, 5.20, 6.1]
Xnew_reshaped = np.array(Xnew).reshape(-1,1)

#--------
# Step 10: Make new prediction
yhat = model.predict(Xnew_reshaped)

#--------
# Step 11: Display XNew and Predicted values
myList = {"XNew":Xnew_reshaped[:,0], "Predictedy":yhat[:,0]}
df_prediction = pd.DataFrame(myList)
df_prediction.head()
```
