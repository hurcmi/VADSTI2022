```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
base_url  = 'https://en.wikipedia.org/wiki/Barack_Obama'
```

```python
resp = requests.get(base_url)
```

```python
resp.ok
```

```python
resp.status_code
```

```python
obama_ht = resp.content
```

```python
obama_ht
```

```python
obama_soup = bs(obama_ht, 'lxml')
```

```python
#obama_soup
```

```python
with open('Obama_wikipedia.html','wb') as file:
    file.write(obama_soup.prettify('utf-8'))
```

```python
div_achi = obama_soup.find_all('div',{'role':'note'})
```

```python
div_achi
```

```python
achi_link =[div.find('a') for div in div_achi]
```

```python
achi_link
```

```python
len(achi_link)
```

```python
div_link = []
for oba_link in div_achi:
    anchor = oba_link.find_all('a')
    for a in anchor:
        div_link.append(a)
```

```python
len(div_link)
```

```python
div_link
```

```python
# Next step is extract all the url for this list
```

```python
relative_link = [l.get('href') for l in div_link]
```

```python
relative_link
```

```python
from urllib.parse import urljoin
```

```python
full_url = [urljoin(base_url,url) for url in relative_link]
```

```python
full_url
```

```python
obama_para = []
i = 0
for url in full_url:
    div_response =requests.get(url)
    print('URL#{0}:{1}'.format(i+1,url))
    div_html = div_response.content
    div_soup = bs(div_html, 'lxml')
    achi_par = div_soup.find_all('p')
    achi_text = [p.text for p in achi_par]
    obama_para.append(achi_text)
    i += 1
```

```python
obama_para[2]
```

```python
all_obama_para = [" ".join(text) for text in obama_para]
```

```python
all_obama_para_achi = dict(zip(full_url, all_obama_para))
```
