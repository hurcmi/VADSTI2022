```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://en.wikipedia.org/wiki/Frederick_Douglass'
```

```python
response = requests.get(url)
```

```python
response.ok
```

```python
response.status_code
```

```python
soup = bs(response.text, 'lxml')
```
